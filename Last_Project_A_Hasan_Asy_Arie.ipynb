#**Unemployment and Mental Illness Survey**

Dataset ini adalah sebuah data yang berisikan hasil survey terkait tingkat pengangguran yang tinggi diantara orang-orang yang memiliki penyakit mental

## Pertanyaan Analisis

Apakah semakin tinggi tingkat pendidikan akan memengaruhi semakin tingginya penyakit mental?

## Data Preparation

```
import pandas as pd
import plotly.express as px
```

```
df = pd.read_excel('/content/Mental Illness Survey 1.xlsx')
df.head()
```

```
df.info()
```

```
data = df.rename({'I identify as having a mental illness':'Mental Illness'},axis=1)
data.head()
```

```
data[['Mental Illness', 'Education']].head()
```

```
list(pd.unique(data['Education']))
```

```
data['Education'].value_counts()
```

## Data Visualization

```
bar = px.bar(data, 
             x = 'Education', 
             y = 'Mental Illness')                          
bar.update_layout(
    title = 'Pengaruh Tingkat Pendidikan terhadap Tingginya Tingkat penyakit Mental',
    title_font_color = 'blue',
    title_font_size = 20,
    xaxis = dict(),
    yaxis = dict(title = '')
    )

bar.show()
```
